```python
def check_passport(passport):
    bypass = ('cid' not in passport and len(passport)>= 14) or len(passport) == 16
    if not bypass:
        return False
    else:
        print(passport)
        check_eyr=len(passport[passport.index('eyr')+1])==4 and 2020<= int(passport[passport.index('eyr')+1]) <= 2030
        check_hgt= (150<=int(passport[passport.index('hgt')+1][:-2])<=193 and passport[passport.index('hgt')+1][-2:]=='cm') or (59<=int(passport[passport.index('hgt')+1][:-2])<=76 and passport[passport.index('hgt')+1][-2:]=='in') if passport[passport.index('hgt')+1][-2:] in ['in', 'cm'] else False
        check_hcl= passport[passport.index('hcl')+1][0]=='#' and len(passport[passport.index('hcl')+1]) ==7 and all([char.isnumeric() for char in passport[passport.index('hcl')+1][1:] if char not in ['a', 'b', 'c', 'd', 'e', 'f']])
        check_ecl= passport[passport.index('ecl')+1] in ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']
        check_byr=len(passport[passport.index('byr')+1])==4 and 1920<= int(passport[passport.index('byr')+1]) <= 2002
        check_iyr=len(passport[passport.index('iyr')+1])==4 and 2010<= int(passport[passport.index('iyr')+1]) <= 2020
        check_pid=len(passport[passport.index('pid')+1])==9
    #print(passport,{'eyr':check_eyr, 'hcl':check_hcl, 'ecl':check_ecl, 'byr':check_byr,'iyr': check_iyr,'pid': check_pid, 'hgt':check_hgt})
    return all([check_eyr, check_hcl, check_ecl, check_byr, check_iyr, check_pid, check_hgt])
```
